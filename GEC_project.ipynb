```python
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_excel("/content/sample_data/(Cleaned) GEC _ Cape Town _ Data Analytics Project (1).xlsx")
ds
```

```python
# prompt: build a line graph with x as the host company and y as the number of placements

plt.figure(figsize=(15, 7))
sn.lineplot(x = "Country/Region", y = "Number of Employees", data=ds)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sn.lineplot(x = "Host Company", y = "Number of placements", data=ds)
plt.xticks(rotation=90)
plt.show()
```

```python
# prompt: do a pivot table and a maximum number of placement per company and tabulate the information

pivot_table = pd.pivot_table(ds, values="Number of Employees", index="Host Company", aggfunc=np.max)
print(pivot_table)
```

```python
count = 0
for indx , row in pivot_table.iterrows():
  if int (row["Number of Employees"]) < 50:
    count+=1
print("We have " + str(count) + " Small Businesses")
count = 0
for indx , row in pivot_table.iterrows():
  if int (row["Number of Employees"]) > 49 and int(row["Number of Employees"])<249 :
    count+=1
print("We have " + str(count) + " Medium Businesses")
count = 0
for indx , row in pivot_table.iterrows():
  if int (row["Number of Employees"]) > 249 :
    count+=1
print("We have " + str(count) + " Large Businesses")
```
