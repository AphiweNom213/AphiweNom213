```python
import requests
from bs4 import BeautifulSoup

URL = "https://en.wikipedia.org/wiki/Johnnie_Cochran#Illness_and_death"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

results = soup.find(id = "mw-page-container-inner")
print(soup.text)
```
